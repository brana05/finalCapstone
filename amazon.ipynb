```python
#importing Pandas
import pandas as pd 
```

```python
#loading data frame 
df=pd.read_csv('amazon_product_reviews.csv')
#present some lines of data
df.head()
```

```python
#present columns contained in the data set
df.columns
```

```python
#give more general info about dataset 
df.info
```

```python
#I will select two relevant columns 
subset_columns=['name','reviews.text','reviews.rating']
df_subset=df[subset_columns].copy()
df_subset.head
```

```python
#Load all nan containing data lines in df
nan_columns=df_subset.isnull().any()
#calculate how many nan containing lines are present in the data set
nan_number=nan_columns.sum()
print(f'Number of nan lines :{nan_number}')
#remove all of the lines with nan if needed 
if nan_number>0:
    df_sub_no_nan=df_subset.dropna()
# if number of NaN ==0 do nothing 
else:
    df_sub_no_nan=df_subset.copy()
df_sub_no_nan.head
```
